```python
import os
import zipfile
import shutil

# Company list
companies = [
    "Collins Aerospace",
    "Airbus",
    "IndiGo",
    "Vistara - Tata SIA Airlines Ltd",
    "United Airlines",
    "British Airways",
    "Air India Limited",
    "Jet Airways",
    "Qatar Airways",
    "Emirates",
    "Volvo Group",
    "Hero MotoCorp",
    "Varroc",
    "JCB",
    "JLR",
    "Toyota Motor Corporation",
    "Cummins India",
    "JSW",
    "JSW Power",
    "Apollo Tyres Ltd",
    "Renault Nissan Technology & Business Centre India",
    "Mahindra Logistics",
    "Tata Technologies",
    "Stellantis",
    "Tech Mahindra",
    "TVS Supply Chain Solutions",
    "Endurance Technology Ltd",
    "JSW Energy Ltd",
    "Worley",
    "Siemens",
    "Xperi Inc",
    "HTIC Global",
    "Segula Technologies",
    "Knorr-Bremse Technology Center India Private Limited (TCI)",
    "Systra",
    "Siemens Energy",
    "Blue Dart",
    "Alstom",
    "CEVA",
    "Wheelseye",
    "UPS",
    "FedEx",
    "Xpressbees",
    "RXO, Inc",
    "Siemens Logistics",
    "DHL Express India",
    "Hapag-Lloyd",
    "Continental",
    "Tata Elxsi",
    "Linde",
    "Hitachi",
    "Jindal Steel and Power",
    "BHP",
    "Aramco",
    "Shell",
    "Indegene",
    "IQVIA",
    "GE Healthcare",
    "NielsenIQ",
    "Baxter International",
    "Eli Lilly and Company",
    "CareStack",
    "Procter and Gamble",
    "Tata Communications"
]

# Description
Description = [

    "is an outstanding pioneer in integrated aerospace and defense solutions, renowned for its unwavering commitment to innovation and reliability.",
    "is a trailblazing aircraft manufacturer celebrated for its visionary designs, exceptional safety standards, and sustainable aviation leadership.",
    "is India’s most admired low-cost carrier, lauded for its flawless punctuality, customer-centric service, and expansive domestic footprint.",
    "is a distinguished full-service airline in India, revered for its graceful hospitality, premium in-flight experience, and immaculate attention to detail.",
    "is a powerhouse of global air travel, celebrated for its extensive route network, award‑winning loyalty program, and commitment to passenger comfort.",
    "is the epitome of British elegance in aviation, famed for its refined service, global reach, and sterling reputation for reliability.",
    "is India’s flagship airline, hailed for its legacy of national service, expanding global footprint, and renewed commitment to excellence.",
    "is a reimagined aviation brand striving to set new benchmarks in premium air travel and customer delight.",
    "is a global aviation icon, known for redefining luxury, excellence in service, and industry-leading in-flight experiences.",
    "is a world-renowned airline synonymous with grandeur, impeccable hospitality, and a vast international network.",
    "is a global engineering legend in transport and infrastructure, admired for its durability, sustainability, and precision manufacturing.",
    "is the undisputed leader in the two-wheeler segment, recognized for empowering millions with affordable, efficient, and innovative mobility solutions.",
    "is a trusted automotive components giant, known for its cutting-edge technology, dependable quality, and global customer base.",
    "is a hallmark of heavy machinery excellence, applauded for its robust designs, construction innovation, and industry leadership.",
    "is a premium automobile brand, blending British luxury with advanced technology to craft world-class vehicles of distinction.",
    "is the world’s most reliable automaker, praised for its commitment to quality, sustainability, and continuous innovation.",
    "is a titan in power solutions, known for its engineering excellence, strong market presence, and customer-first approach.",
    "is a dynamic industrial conglomerate, recognized for redefining India's infrastructure through steel, energy, and innovation.",
    "is a robust player in India's power sector, noted for its focus on clean energy, operational efficiency, and strategic growth.",
    "is a global tyre manufacturer, lauded for its breakthrough designs, high-performance products, and sustainable practices.",
    "is a technological nerve center for Renault-Nissan, driving future mobility through smart engineering and collaborative innovation.",
    "is a key logistics force in India, respected for its scalable solutions, timely execution, and robust operational backbone.",
    "is a digital engineering powerhouse, known for enabling global transformation through product development and technology services.",
    "is an automotive behemoth, crafting forward-thinking vehicles with a strong legacy of excellence and innovation.",
    "is a digital transformation pioneer, redefining the global IT landscape with agility, vision, and technological depth.",
    "is a trusted leader in supply chain innovation, offering seamless, data-driven solutions across industries.",
    "is a top-tier auto component manufacturer, acclaimed for its engineering prowess, efficiency, and customer satisfaction.",
    "is a key driver in India's energy transition, renowned for operational integrity and sustainable power generation.",
    "is a globally respected engineering consultancy, delivering cutting-edge solutions for the energy, chemicals, and resources sectors.",
    "is a global leader in electrification and automation, setting new standards in smart infrastructure and digital industries.",
    "is a frontier technology company, renowned for enhancing digital experiences through audio, imaging, and AI-driven innovations.",
    "is a fast-emerging healthcare innovator, delivering transformative digital solutions with speed and precision.",
    "is a global engineering leader, celebrated for shaping the future of mobility, energy, and industrial solutions.",
    "is a global hub of braking system innovation, empowering transport safety and efficiency through next-gen technology.",
    "is a visionary in urban transport design, known for engineering intelligent, sustainable mobility across continents.",
    "is a bold force in the global energy space, driven by clean technology, reliability, and climate consciousness.",
    "is India’s premier express logistics provider, trusted for its unmatched speed, wide coverage, and technological edge.",
    "is a world-renowned leader in rail infrastructure, praised for its commitment to green mobility and transportation excellence.",
    "is a logistics titan, empowering global trade through scalable, intelligent, and seamless supply chain solutions.",
    "is a disruptive force in India's trucking industry, leveraging data and technology to create smarter transport networks.",
    "is a global logistics leader, admired for delivering excellence, innovation, and trust in every shipment.",
    "is a courier and logistics powerhouse, celebrated for fast, dependable, and customer-first delivery services.",
    "is a rising star in e-commerce logistics, delivering growth through speed, technology, and reach.",
    "is a next-generation freight brokerage innovator, redefining logistics through intelligent digital platforms.",
    "is a smart logistics integrator, transforming global automation and supply chain ecosystems with German precision.",
    "is a trusted express delivery partner, known for seamless global connectivity, speed, and excellence.",
    "is a maritime logistics giant, moving global trade forward with efficiency, trust, and innovation.",
    "is an innovation-driven automotive leader, shaping the future of mobility with intelligent systems and sustainability.",
    "is a premium design and technology company, empowering automotive, media, and healthcare industries worldwide.",
    "is a global industrial gases leader, driving innovation in sustainability, clean energy, and productivity.",
    "is a global technology powerhouse, leading the charge in infrastructure, mobility, and advanced electronics.",
    "is a vital pillar of India’s steel industry, fueling national development through innovation and strength.",
    "is a global mining leader, acclaimed for sustainable practices, operational excellence, and resource stewardship.",
    "is the world’s largest energy company, renowned for its scale, innovation, and contribution to global energy security.",
    "is a global energy leader, committed to advancing cleaner fuels, innovation, and a sustainable energy future.",
    "is a digital transformation partner in healthcare, revolutionizing life sciences through data, platforms, and innovation.",
    "is a global leader in healthcare data analytics and research, driving precision medicine and evidence-based decisions.",
    "is a medical technology pioneer, empowering care providers with breakthrough imaging, monitoring, and diagnostics.",
    "is a top-tier consumer intelligence firm, helping businesses grow through data, insight, and foresight.",
    "is a global healthcare company known for lifesaving technologies and a legacy of improving patient care.",
    "is a pharmaceutical leader, acclaimed for groundbreaking therapies and a deep commitment to global well-being.",
    "is a cloud-based platform revolutionizing dental care through modern, integrated, and scalable practice solutions.",
    "is a global consumer goods giant, loved for its trusted brands, quality products, and commitment to sustainability.",
    "is a digital communications leader, enabling global businesses with robust, secure, and transformative connectivity."

]

# Paths
template_path = "Invitation_IE.docx"
output_folder = "Invitation_Industrial_Core_Company_list"
os.makedirs(output_folder, exist_ok=True)

def generate_personalized_doc(company_name,Description, index):
    # temp folder for unzip/edit
    temp_folder = f"temp_{index}"
    if os.path.exists(temp_folder):
        shutil.rmtree(temp_folder)
    os.makedirs(temp_folder)

    with zipfile.ZipFile(template_path, 'r') as zip_ref:
        zip_ref.extractall(temp_folder)

    # document.xml (main body)
    doc_path = os.path.join(temp_folder, "word", "document.xml")
    with open(doc_path, "r", encoding="utf-8") as file:
        content = file.read()

    # replace placeholder
    content = content.replace("[Company Name]", company_name)
    content = content.replace("[Description]", description)
    
    # write back modified content
    with open(doc_path, "w", encoding="utf-8") as file:
        file.write(content)

    # create new docx
    output_docx = os.path.join(output_folder, f"Invitation_{index}_{company_name.replace(' ', '_')}.docx")
    shutil.make_archive("temp_doc", 'zip', temp_folder)
    shutil.move("temp_doc.zip", output_docx)

    # delete temp
    shutil.rmtree(temp_folder)

for idx, (company, description) in enumerate(zip(companies, Description), 1):
    generate_personalized_doc(company, description, idx)

shutil.make_archive('Invitation_Industrial_Core_Company_list', 'zip', 'Invitation_Industrial_Core_Company_list')
print(f"All invitations created.")
```
